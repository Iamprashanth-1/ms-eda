```python
import pandas as pd
```

```python
df=pd.read_csv('Datas1.csv')
```

```python
df.head()
```

```python
sd=df.sum(axis = 0, skipna = True)
sd=sd[1:]
```

```python
dat=[]
valu=[]
for i,v in sd.items():
    dat.append(i)
    valu.append(v)
```

```python


import plotly.graph_objects as go
from collections import Counter
import datetime
import pandas as pd
import plotly.express as px
import dash_bootstrap_components as dbc
import dash
import dash_core_components as dcc
import dash_html_components as html

dif=dict()
sif=[]
dif['Date']=dat
dif['count']=valu
sif.append(dif)
fig7 =px.line(dif,x='Date',y='count',title='Time Series with Range Slider and Selectors')

fig7.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)


fig7.update_layout(
    height=600,
    template="plotly_dark",

)
fig7.update_xaxes(rangeslider_visible=True)
```

```python
ds=list(df)
ds=ds[1:]
print(ds[-1])
```

```python
cv=[]
csx=[]
for i in range(len(ds)-1):
    p=ds[i][3:5]
    if p==ds[i+1][3:5]:
        cv.append(ds[i])
    else:
        cv.append(ds[i])
        #print(cv)
        csx.append(cv)
        cv=[]
csx.append(cv)

```

```python
print(csx)
```

```python
fgs=[]
for i in csx:
    gv=str(i[0][3:])
    fgs.append(gv)
    df[gv]=df[i].sum(axis=1)
```

```python
df.sort_values(by =fgs,ascending=True)
```
